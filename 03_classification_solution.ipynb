````python
numericals = list(df.dtypes[df.dtypes != 'object'].index)
numericals
````

````python
corr_dict = {}
for i in numericals:
    for j in numericals:
        if i != j:
            corr = df[i].corr(df[j])
            corr_dict[f'{i} and {j} coorrelation is: '] = corr
        
````

````python
# Find the key with the maximum value in the dictionary
max_key = max(corr_dict, key=corr_dict.get)

# Get the maximum value
max_value = corr_dict[max_key]

# Print the key and the value
print(f'The highest correlation is {max_key} with a value of {max_value}')
````

### Question 5 

* Let's find the least useful feature using the *feature elimination* technique.
* Train a model with all these features (using the same parameters as in Q4).
* Now exclude each feature from this set and train a model without it. Record the accuracy for each model.
* For each feature, calculate the difference between the original accuracy and the accuracy without the feature. 

Which of following feature has the smallest difference?

- `age`
- `balance`
- `marital`
- `previous`

> **Note**: The difference doesn't have to be positive.

````python
df.head()
````

````python
all_features = categoricals + numericals
omitt = ['age', 'balance', 'marital', 'previous']
````

````python
original_accuracy = (y_val == subscribe_decision).mean()

fe_dictionary = {}

for i in omitt:
    df_train_omitt = df_train.copy()
    df_val_omitt = df_val.copy()
    del df_train_omitt[i]
    del df_val_omitt[i]
    train_dicts = df_train_omitt.to_dict(orient= 'records')
    X_train = dv.fit_transform(train_dicts)
    val_dicts = df_val_omitt.to_dict(orient= 'records')
    X_val = dv.transform(val_dicts)
    model.fit(X_train, y_train)
    y_pred = model.predict_proba(X_val)[:, 1]
    subscribe_decision = (y_pred > 0.5)
    accuracy = (y_val == subscribe_decision).mean()
    print(f'The accuracy without {i} feature is {accuracy} compare to {original_accuracy} which is the "original accuracy"')
    print()
    decrease_in_accuracy = original_accuracy - accuracy
    fe_dictionary[i] = decrease_in_accuracy

sorted_dict = dict(sorted(fe_dictionary.items(), key=lambda item: item[1], reverse=True))
sorted_dict
    
````

### Question 6

* Now let's train a regularized logistic regression.
* Let's try the following values of the parameter `C`: `[0, 0.01, 0.1, 1, 10]`.
* Train models using all the features as in Q4.
* Calculate the accuracy on the validation dataset and round it to 3 decimal digits.

Which of these `C` leads to the best accuracy on the validation set?

- 0.01
- 0.1
- 1
- 10
- 100

> **Note**: If there are multiple options, select the smallest `C`.

````python
original_accuracy = (y_val == subscribe_decision).mean()

regs = [0.001, 0.01, 0.1, 1, 10]

reg_dictionary = {}


for i in regs:
    model = LogisticRegression(solver='liblinear', C=i, max_iter=1000, random_state=42)
    df_train = df_train.copy()
    df_val = df_val.copy()
    train_dicts = df_train[categoricals + numericals].to_dict(orient= 'records')
    X_train = dv.fit_transform(train_dicts)
    val_dicts = df_val[categoricals + numericals].to_dict(orient= 'records')
    X_val = dv.transform(val_dicts)
    model.fit(X_train, y_train)
    y_pred = model.predict_proba(X_val)[:, 1]
    subscribe_decision = (y_pred > 0.5)
    accuracy = (y_val == subscribe_decision).mean()
    print(f'The accuracy with {i} regularization is {accuracy}')
    
    reg_dictionary[i] = accuracy.round(3)


sorted_reg_dict = dict(sorted(reg_dictionary.items(), key=lambda item: item[1], reverse=True))
sorted_reg_dict
````
